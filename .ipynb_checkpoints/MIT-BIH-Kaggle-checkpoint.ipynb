# Data Description
The MIT-BIH Arrhythmia Database contains 48 half-hour excerpts of two-channel ambulatory ECG recordings, obtained from 47 subjects studied by the BIH Arrhythmia Laboratory between 1975 and 1979. Twenty-three recordings were chosen at random from a set of 4000 24-hour ambulatory ECG recordings collected from a mixed population of inpatients (about 60%) and outpatients (about 40%) at Boston's Beth Israel Hospital; the remaining 25 recordings were selected from the same set to include less common but clinically significant arrhythmias that would not be well-represented in a small random sample.

The recordings were digitized at **360 samples per second** per channel with 11-bit resolution over a 10 mV range. Two or more cardiologists independently annotated each record; disagreements were resolved to obtain the computer-readable reference annotations for each beat (approximately 110,000 annotations in all) included with the database.

The database contains **23 records** (**numbered from 100 to 124** inclusive with some numbers missing) chosen at random from a set of over 4000 long-term Holter recordings, and **25 records** (**numbered from 200 to 234** inclusive, again with some numbers missing) selected from the same set to include a variety of rare but clinically important phenomena that would not be well-represented by a small random sample of Holter recordings. Each of the 48 records is slightly over 30 minutes long.

## ECG lead configuration
In most records, the upper signal is a modified limb lead II (MLII), obtained by placing the electrodes on the chest. The lower signal is usually a modified lead V1 (occasionally V2 or V5, and in one instance V4); as for the upper signal, the electrodes are also placed on the chest. This configuration is routinely used by the BIH Arrhythmia Laboratory. Normal QRS complexes are usually prominent in the upper signal. The lead axis for the lower signal may be nearly orthogonal to the mean cardiac electrical axis, however (i.e., normal beats are usually biphasic and may be nearly isoelectric). Thus normal beats are frequently difficult to discern in the lower signal, although ectopic beats will often be more prominent (see, for example, record 106). A notable exception is record 114, for which the signals were reversed. Since this happens occasionally in clinical practice, arrhythmia detectors should be equipped to deal with this situation. In records 102 and 104, it was not possible to use modified lead II because of surgical dressings on the patients; modified lead V5 was used for the upper signal in these records.

## Plot the distribution graph of the columns (signals) of the recorded data.

Get DataFrame as input and print the distribution plot (**histogram**) of the upper signal column ,for example **'MLII'**, and lower signal column ,for example **'V5'**.

Also, print the number of unique values of each column.

```python
# Distribution graphs (histogram/bar graph) of column data
def plotPerColumnDistribution(df, nGraphShown, nGraphPerRow):
    nunique = df.nunique()
    # For displaying purposes, pick columns that have between 1 and 500 unique values
    df = df[[col for col in df if nunique[col] > 1 and nunique[col] < 500]]
    nRow, nCol = df.shape
    columnNames = list(df)
    nGraphRow = (nCol + nGraphPerRow - 1) / nGraphPerRow
    plt.figure(num = None, figsize = (6 * nGraphPerRow, 3 * nGraphRow), 
               dpi = 80, facecolor = 'w', edgecolor = 'k')
    for i in range(min(nCol, nGraphShown)):
        plt.subplot(nGraphRow, nGraphPerRow, i + 1)
        columnDf = df.iloc[:, i]
        if (not np.issubdtype(type(columnDf.iloc[0]), np.number)):
            valueCounts = columnDf.value_counts()
            valueCounts.plot.bar()
        else:
            columnDf.hist(bins=100)
        plt.ylabel('counts')
        plt.xticks(rotation = 90)
        plt.title(f'{columnNames[i]} (column {i})')
    plt.tight_layout(pad = 1.0, w_pad = 1.0, h_pad = 1.0)
    plt.show()
    for i in range(nCol):
        print("The number of unique values for the column", columnNames[i], " = ", nunique[i+1])
```

There is multiple correlation methods I tested multiple correlations and plot them all in one figure in the bellow

```python
def histogram_intersection(a, b):
    v = np.minimum(a, b).sum().round(decimals=1)
    return v

# Correlation matrix
def plotCorrelationMatrix(df, graphWidth):
    filename = df.dataframeName
    df = df.dropna('columns') # drop columns with NaN
    # keep columns where there are more than 1 unique values
    df = df[[col for col in df if df[col].nunique() > 1]]
    if df.shape[1] < 2:
        print(f'No correlation plots shown: The number of non-NaN',
              ' or constant columns ({df.shape[1]}) is less than 2')
        return
    plt.figure(num = None, figsize = (graphWidth, graphWidth), 
               dpi = 80, facecolor = 'w', edgecolor = 'k')
    corr_methods = ['pearson', 'spearman', 'kendall', histogram_intersection]
    for i in range(4):
        plt.subplot(2, 2, i+1)
        corr = df.corr(method=corr_methods[i])
        plt.matshow(corr, fignum = False)
        plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
        plt.yticks(range(len(corr.columns)), corr.columns)
        plt.gca().xaxis.tick_bottom()
        plt.colorbar()
        if (i==3):
            plt.title(f' (histogram_intersection) Correlation Matrix for {filename}', fontsize=8)
        else:
            plt.title(f' ({corr_methods[i]}) Correlation Matrix for {filename}', fontsize=10)
        
    plt.tight_layout(pad = 1.0, w_pad = 1.0, h_pad = 1.0)
    plt.show()
```

The previous plot was in the domain of **Sample Number** and I want to show it in the time domain as the *sample rate* is **360 samples per second**, I will convert the sample number to time(second).

```python
# Convert sample # to the time formate (Vectorize implementation for speed-up)
def convert_sample_to_time(sample):
    sec, m_sec = np.divmod(sample, 360)
    m_sec = np.round(m_sec*100/36)
    mon, sec = np.divmod(sec, 60)
    
    mon_str = np.char.mod('%02d', mon)
    sec_str = np.char.mod('%02d', sec)
    m_sec_str = np.char.mod('%03d', m_sec)
    
    conv_time_str = np.char.add(mon_str, ":")
    conv_time_str = np.char.add(conv_time_str, sec_str)
    conv_time_str = np.char.add(conv_time_str, ".")
    conv_time_str = np.char.add(conv_time_str, m_sec_str)
    
    return mon, sec, m_sec, conv_time_str
```

```python
def ECG_plot_time_domain(df):
    samples = df[df.columns[0]]
    (_, _, _, sample_to_time_str) = convert_sample_to_time(samples)
    lower = df[df.columns[2]]
    upper = df[df.columns[1]]

    fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(17,5))
    ax.plot(sample_to_time_str,lower, ls='-', label=df.columns[2])
    ax.plot(sample_to_time_str,upper, 'r', ls='-', label=df.columns[1])
    ax.set_ylabel('Millivolts (mV)', fontsize=15)
    ax.set_xlabel('Time', fontsize=15)
    ax.set_title(f'Both upper and lower signal in ((Time)) domain'
                , fontsize=15)


    # Major ticks every 90 samples 0.25 sec , minor ticks every 45 samples 0.125 sec
    major_ticks_x = np.arange(-20, 1040, 90)
    minor_ticks_x = np.arange(-20, 1040, 45)

    ax.set_xticks(major_ticks_x)
    ax.set_xticks(minor_ticks_x, minor=True)
    
    # And a corresponding grid
    ax.grid(which='both')

    # Or if you want different settings for the grids:
    ax.grid(which='minor', alpha=0.2)
    ax.grid(which='major', alpha=0.5)  

    ax.legend(framealpha=0.6, loc=1)
    plt.show()
```

### Make the plots scrollable and interactive
After plotting only 1000 sample points of the ECG signal I want to make it possible to interactively I scroll through the singal. As each signal is about 30 minutes of the recorded data with the sample rate of 360 per second, I am making it interactive to scroll the signal to the right and left by using *ipywidgets*.

```python
from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
from ipywidgets import Layout


def read_data_df(name='100', nRowsRead = None):
    # nRowsRead = None # specify 'None' if want to read whole file
    # 100.csv may have more rows in reality, but we are only loading/previewing the first 1000 rows
    data_df = pd.read_csv('./mitbih-database/'+name+'.csv', delimiter=',', nrows = nRowsRead)
    data_df.dataframeName = name+'.csv'
    return data_df

def read_annotation_df(name='100'):
    annotations = pd.read_csv('./mitbih-database/'+name+'annotations.txt', delim_whitespace=True )
    annotations.drop(['Aux'], axis=1, inplace=True)
    annotations.columns = ["Time", "Sample", "Type", "Sub", "Chan", "Num", "Aux"]
    return annotations

def plot_pagination(data_df ,length_of_plot=1000, page=2):
    paged_data_df = data_df[(page-1)*length_of_plot : page*length_of_plot]
    return paged_data_df

def number_of_pages(data_df ,length_of_plot=1000):
    pages_count = data_df.shape[0]//length_of_plot
    return pages_count

data_df = read_data_df(name='100')
annotations_df = read_annotation_df(name='100')
pages_count = number_of_pages(data_df ,length_of_plot=1000)

def scrollable_ECG_time_domain_annotated(page_slide):
    paged_data_df = plot_pagination(data_df ,length_of_plot=1000, page=page_slide)
    ECG_plot_time_domain_annotated(paged_data_df, annotations_df, length_of_plot=1000, page=page_slide)
    return

int_slide=widgets.IntSlider(min=1, 
                            max=pages_count, 
                            step=1, 
                            value=1, 
                            continuous_update=False, 
                            layout=Layout(width='95%'))
previous_botton = widgets.Button(
    description='Previous',
    disabled=False,
    button_style='info',
    tooltip='Previous',
    icon='arrow-left' 
)

next_botton = widgets.Button(
    description='Next',
    disabled=False,
    button_style='info',
    tooltip='Next',
    icon='arrow-right' 
)


out = widgets.interactive_output(scrollable_ECG_time_domain_annotated, {'page_slide': int_slide})
ui = widgets.HBox([previous_botton, int_slide, next_botton])

display(ui, out)

# set event handlers for next and previous buttons
def change_page(button):
    if button.description=='Previous':
        if (int_slide.value-1 ==0):
            pass
        else:
            int_slide.value -= 1

    elif button.description=='Next':
        if (int_slide.value+1 == (int_slide.max + 1)):
            pass
        else:
            int_slide.value += 1

next_botton.on_click(change_page)
previous_botton.on_click(change_page)
```

#### Detecting R-peaks in ECG signal:

To detect R-peaks in the signal, I do the following steps:
1. Specify the Region Of Interest (**ROI**) for each R-peak.
2. Determine **R-peak** (the maximum of the points) in the region of interest.

**What is ROI**: This area basically is the area of the signal which we want to find the arg_maximum of it.
In other words, R-peak is a maximum. To find each maximum point, I first find the area in which the maximum point is falling in and then I find the point with the maximum value.

**How to determine R-peak**: there are three suggested way to find the R-peak in the ROI.
- Fit a curve or interpolate a curve to the points in the ROI and find the arg_max of the curve.
- Find the slopes between each two adjacent points and detect the point where slope is reversed.
- Mark points within ROI, Find position of the highest point.

I am using the last and simplest approach which is just find the position of the highest point.

#### 1. How I specify ROI in my ECG signal for finding R-peak?
For finding this region, I first compute the **simple moving average** of the ECG signal then I took the areas in which the ECG signal value is higher than the simple moving average. This area is my ROI (Region Of Interest).

#### 2. How I determine R-peaks given my RIO?
I simply taking the datapoint which have the maximum value in the ROI.

### Implementation:
I first writing some helper functions to divide and conquer the algorithm to finding R-peaks.

#### Read data:
Read only first 200000 datapoints for simplicity

```python
def read_data_df(name='100', nRowsRead=None):
#     nRowsRead = None # specify 'None' if want to read whole file
    # 100.csv may have more rows in reality, but we are only loading/previewing the first 1000 rows
    data_df = pd.read_csv('./mitbih-database/'+name+'.csv', delimiter=',', nrows = nRowsRead)
    data_df.dataframeName = name +'.csv'
    return data_df
```

#### Simple moving average:
Compute simple moving average of the signal using the window size of 0.75s.

```python
# simple moving average
def movingaverage(original_ecg, hrw=0.75, fs = 360):
    #Calculate moving average with 0.75s in both directions
    hrw = 0.75 #One-sided window size, as proportion of the sampling frequency
    fs = 360 #The example dataset was recorded at 360Hz
    mov_avg = original_ecg.rolling(int(hrw*fs)).mean() #Calculate moving average
    
    #Impute where moving average function returns NaN, which is the beginning of the signal where x hrw
    avg_hr = (np.mean(original_ecg))
    mov_avg = [avg_hr if math.isnan(x) else x for x in mov_avg]
    return mov_avg
```

#### Detect R-peaks:
Using the computed moving average, I detect R-peaks. In simple terms, I take the area of the ECG signal which is on top of the moving average line and this area is my ROI and then I get the maximum datapoint of this area as the R-peak of the beat.

However, for better detection I shifted the moving average line to some percentage to the top. Later, I will test different values for **ma_perc** to find the best shift of the moving average line for better R-peak detection.

```python
def detect_peaks(original_ecg, mov_avg, ma_perc):
    #Raise the average by ma_perc% to prevent the secondary heart contraction from interfering
    mov_avg = [(x+((x/100)*ma_perc)) for x in mov_avg]

    #Mark regions of interest
    window = []
    peaklist = []
    
    #We use a counter to move over the different data columns
    listpos = 0
    for datapoint in original_ecg:
        rollingmean = mov_avg[listpos] #Get local mean

        #If no detectable R-complex activity -> do nothing
        if (datapoint <= rollingmean) and (len(window) < 1):
            listpos += 1
        #If signal comes above local mean, mark ROI
        elif (datapoint > rollingmean):
            window.append(datapoint)
            listpos += 1
        #If signal drops below local mean -> determine highest point
        else:
            maximum = max(window)
            #Notate the position of the point on the X-axis
            beatposition = listpos - len(window) + (window.index(max(window)))
            peaklist.append(beatposition) #Add detected peak to list
            window = [] #Clear marked ROI
            listpos += 1
    #Get the y-value of all peaks for plotting purposes
    ybeat = [original_ecg[x] for x in peaklist]
    
    return peaklist, ybeat
```

#### Compute some helpful measurements:
From the signal processing point of view I computed some property of the signal.

- **measure_rr**: Compute the distance between each two adjacent R-peaks in Millisecond and return a list (RR_list).
- **measure_bpm**: Compute the Beat Per Minute (BPM) values from the list of (RR_list).
- **measure_bpm_signal**: For plotting purpose, I create a bpm_signal list which is as same size as the original signal.

```python
def measure_rr(peaklist, fs = 360):
    RR_list = []
    cnt = 0
    while (cnt < (len(peaklist)-1)):
        #Calculate distance between beats in # of samples
        RR_interval = (peaklist[cnt+1] - peaklist[cnt])
        #Convert sample distances to ms distances
        ms_dist = ((RR_interval / fs) * 1000.0)
        RR_list.append(ms_dist) #Append to list
        cnt += 1
    return RR_list

def measure_bpm(RR_list):   
    # bpm for each beat.
    bpm_list = [60000/RR_ms_dist for RR_ms_dist in RR_list]
    return bpm_list

def measure_bpm_signal(mov_avg, bpm, peaklist):
    avg_bpm = np.mean(bpm)
    
    bpm_signal = np.repeat(avg_bpm, len(mov_avg))

    i = 0
    last_set_bpm = avg_bpm
    for index, datapoint in enumerate(bpm_signal):
        if (index == peaklist[i]):
            last_set_bpm = bpm[i]
            bpm_signal[index] = last_set_bpm
            if(i < len(bpm)-1):
                i+= 1
        else:
            bpm_signal[index] = last_set_bpm
    return bpm_signal, avg_bpm
```

### Dynamically shitf of moving average for better  R-peak detection.

Here I use the standard deviation of the RR distances (**RRSD**) to select the best shifting threshold for moving average during the R-peak detection. 

Best moving average shift will give us the lowest non-zero RRSD and believable BPM.

First, I compute the standard deviation of the RR distances **RRSD** in function **measure_rrsd()**.

```python
def measure_rrsd(RR_list):
    # list is empty for warning
    if not RR_list:
        return 0
    else:
        return np.std(RR_list)
```

Second, I find the best moving average percentage shift **ma_perc_best** for my ECG data in writing the funtion **find_best_ma_perc_shift**.

I tested very limited options for the **ma_perc_list** for speeding up the code.

```python
def find_best_ma_perc_shift(original_ecg, mov_avg, fs):
    #List with moving average raise percentages, make as detailed as you like but keep an eye on speed
    ma_perc_list = [5, 10, 15, 20, 25, 30]
    rrsd_list = []
    valid_ma = []
    #Detect peaks with all percentages, append results to list 'rrsd'
    for ma_perc in ma_perc_list: 
        (peaklist, ybeat) = detect_peaks(original_ecg, mov_avg, ma_perc)
        peak_bpm = ((len(peaklist)/(len(original_ecg)/fs))*60)
        RR_list = measure_rr(peaklist)
        rrsd = measure_rrsd(RR_list)
        rrsd_list.append([rrsd, peak_bpm, ma_perc])
        
    #Test rrsd_list entries and select valid measures
    # valid moving average percentage shift is the one which give us lowest non-zero RRSD and
    # belivable BPM here I took the belivable BPM (30 < bpm < 130)
    for rrsd, peak_bpm, ma_perc in rrsd_list:
        if ((rrsd > 1) and ((peak_bpm > 30) and (peak_bpm < 130))):
            valid_ma.append([rrsd, ma_perc])
            
    #Save the ma_perc for plotting purposes later on (not needed)
    ma_perc_best = min(valid_ma, key = lambda t: t[0])[1]
    rrsd_best = min(valid_ma, key = lambda t: t[0])[0]
    return ma_perc_best
```

### Running the optimized version of the R-peak detector
After finding the best moving average percentage shift, I find the R-peaks and plot all the analysis again.

```python
def ecg_signal_analysis_optimized(original_ecg, fs = 360):
    mov_avg = movingaverage(original_ecg, hrw=0.75)
    
    # test some possible ma_perc_shift for selecting the best from by looking at RRSD and BPM
    ma_perc_best = find_best_ma_perc_shift(original_ecg, mov_avg, fs)
    
    #Detect peaks with 'ma_perc_best' that goes with lowest rrsd
    (peaklist, ybeat) = detect_peaks(original_ecg, mov_avg, ma_perc = ma_perc_best)
    RR_list = measure_rr(peaklist)
    bpm_list = measure_bpm(RR_list)
    (bpm_signal, avg_bpm) = measure_bpm_signal(mov_avg, bpm_list, peaklist)
    advance_plotting_ecg(original_ecg, mov_avg, bpm_signal, avg_bpm, peaklist, ybeat)
    return

data_df = read_data_df(name='100', nRowsRead = 200000)
original_ecg = data_df[data_df.columns[1]]
ecg_signal_analysis_optimized(original_ecg, fs = 360)
```

### ECG signal noise reduction (Butterworth filter)
Use Butterworth filter to reduce the noise in the ECG signal. But this filter didn't help us at the end.

```python
from scipy.signal import butter, lfilter #Import the extra module required
#Define the filter
def butter_lowpass(cutoff, fs, order=5):
    nyq = 0.5 * fs #Nyquist frequeny is half the sampling frequency
    normal_cutoff = cutoff / nyq
    b, a = butter(order, normal_cutoff, btype='low', analog=False)
    return b, a
def butter_lowpass_filter(data, cutoff, fs, order):
    b, a = butter_lowpass(cutoff, fs, order=order)
    y = lfilter(b, a, data)
    return y
def test_butterworth_filter_ecg(original_ecg):
    #filter the signal with a cutoff at 2.5Hz and a 5th order Butterworth filter
    filtered = butter_lowpass_filter(original_ecg, 2.5, 360, 5)

    fig, (ax1, ax2)  = plt.subplots(nrows=2, ncols=1, figsize=(17,5), sharex=True)
    fig.suptitle('Butterworth Filter on ECG')

    #Plot ECG
    ax1.plot(original_ecg, color='blue', ls='-', label="Original_ECG")
    ax2.plot(filtered, color='Red', label='Filtered Signal')

    ax1.legend(loc=1)
    ax2.legend(loc=1)
    plt.show()
    return

data_df = read_data_df(name='100', nRowsRead=10000)
original_ecg = data_df[data_df.columns[1]]
test_butterworth_filter_ecg(original_ecg)
```

### Splitting signal into Heart Beats:
I do the simple way of splitting. I take the area between each two adjacent R-peak.

First, I am writing a function to split the ECG signal **original_ecg** and its computed measured signal into its heart beats using the index of the R-peaks **peaklist**

```python
def hb_split(original_ecg, peaklist, mov_avg, bpm_signal, annotations_df):
    heart_beats = [original_ecg[i : j] for i, j in zip([0] + peaklist, peaklist + [None])]
    heart_beats = heart_beats[1:-1]
    
    hbs_mov_avg = [mov_avg[i : j] for i, j in zip([0] + peaklist, peaklist + [None])]
    hbs_mov_avg = hbs_mov_avg[1:-1]
    
    hbs_bpm_signal = [bpm_signal[i : j] for i, j in zip([0] + peaklist, peaklist + [None])]
    hbs_bpm_signal = hbs_bpm_signal[1:-1]
    
    hbs_annotations = []
    for hb in heart_beats:
        # get the maximum and minimum of the sample numbers of each heart beat
        max_sample_number = hb.index.max()
        min_sample_number = hb.index.min()

        # sort to get only annotations for the sample numbers in range
        sorted_annotations_df = annotations_df.loc[(annotations_df['Sample'] <= max_sample_number) 
                                                   & (annotations_df['Sample'] >= min_sample_number)]
        hbs_annotations.append(sorted_annotations_df)
    
    return heart_beats, hbs_mov_avg, hbs_bpm_signal, hbs_annotations
```

## Create the visibility graph of each heart beat
After splitting signal to heart beats, I am going to generate the graph of each heart beat of the signal using the Visibility Algorithm.

For building the visibility graph, we do the following steps:
1. We consider every datapoint of the time-series is a node in our visibility graph
2. Every two nodes are connected by an edge if they are visible to each other.

There are two versions of the visibility.

- Natural visibility: two points visible with a line.
- Horizontal visibility: two points visible with a horizontal line.

### **Natural visibility:**
Given time series $y = f(t)$. For each two points $A = (t_a, y_a)$ and $B = (t_b, y_b)$ we know $t_a < t_b$. These points are visible if for every intermediate node $C = (t_c, y_c)$, $t_a < t_c < t_b$, we have the following inequality: $y_c < y_a + (y_b - y_a) (\frac{t_c - t_a}{t_b - t_a})$

In simple terms, this means that if we have two point $A$ and $B$ and point $A$ comes before $B$ ($t_a < t_b$). These two points are naturally visible to each other if and only if every intermediatory node $C$ fall bellow the line that connect these two nodes. If the inequality it true it means that the node in between is bellow line that connect points $A$ and $B$.

Here is the implementation for natural visibility graph which is exact implementation of the algorithm

The code taked from the following paper and modified a little bit.
<a href="https://arxiv.org/pdf/1903.01976.pdf">Spectral Visibility Graphs:
Application to Similarity of Harmonic Signals (20 Jun 2019)</a>

<a href="https://github.com/delialia/vgspectra/blob/master/visibility_algorithms.py">Github repository</a>


```python
def nvg(series, timeLine):
    L = len(series)
    # timeLine is the vector containing the time stamps
    #if timeLine == None: timeLine = range(L)

    # initialise output
    all_visible = []


    for i in range(L-1):
        node_visible = []
        ya = float(series[i])
        ta = timeLine[i]

        for j in range(i+1,L):
            yb = float(series[j])
            tb = timeLine[j]

            yc = series[i+1:j]
            tc = timeLine[i+1:j]

            if all( yc[k] < (ya + (yb - ya)*(tc[k] - ta)/(tb-ta)) for k in range(len(yc)) ):
                node_visible.append(tb)

        if len(node_visible)>0 : all_visible.append([ta, node_visible])

    return all_visible
```

Here is the divide and conquer (D&C) approach for implementing the natural visibility algorithm that again taken from the same repository. 

This implementation is recursive and faster compare to the previous basic implementation of the algorithm and here we can specify the specific section of the time-series to build only the visibility graph for that section of the signal not the whole signal.

Code taken from the same repository but the paper can be found in the following link.

<a href="https://arxiv.org/pdf/1905.03204.pdf">Efficient On-line Computation of Visibility Graphs (8 May 2019)</a>

```python
def nvg_dc(series, timeLine, left, right, all_visible = None):

    #if timeLine == None : timeLine = range(len(series))
    if all_visible == None : all_visible = []

    node_visible = []

    if left < right : # there must be at least two nodes in the time series
        k = series[left:right].index(max(series[left:right])) + left

        # check if k can see each node of series[left...right]
        for i in range(left,right):
            if i != k :
                a = min(i,k)
                b = max(i,k)

                ya = float(series[a])
                ta = timeLine[a]
                yb = float(series[b])
                tb = timeLine[b]
                yc = series[a+1:b]
                tc = timeLine[a+1:b]

                if all( yc[j] < (ya + (yb - ya)*(tc[j] - ta)/(tb-ta)) for j in range(len(yc)) ):
                    node_visible.append(timeLine[i])

        if len(node_visible) > 0 : all_visible.append([timeLine[k], node_visible])

        nvg_dc(series,timeLine,left, k, all_visible = all_visible)
        nvg_dc(series,timeLine, k+1, right, all_visible = all_visible)

    return all_visible
```

### **Horizontal visibility:**
Two points $A = (t_a, y_a)$ and $B = (t_b, y_b)$ that point $A$ comes before $B$ ($t_a < t_b$) are visible if and only if for every node $C$ in between we have the following inequality. $y_c < min(y_a,y_b)$.

In other words, horizontal visibility means that two points can be connected together with an horizontal line without crossing the vertical bar that represent the value of each intermediate node.

The code is taken from this <a href="https://github.com/delialia/vgspectra/blob/master/visibility_algorithms.py">Github repository</a> but I found an **error in the line 23** which was **all** and I changed it to **any**. And it still give me the correct answer but faster that before.

and The cited papers are 

- <a href="https://arxiv.org/pdf/1905.03204.pdf">Efficient On-line Computation of Visibility Graphs (8 May 2019)</a>
- <a href="https://arxiv.org/pdf/1002.4526.pdf">Horizontal visibility graphs: exact results for random time series (24 Feb 2010)</a>

```python
def hvg(series, timeLine):
    # series is the data vector to be transformed
    #if timeLine == None: timeLine = range(len(series))
    # Get length of input series
    L = len(series)
    # initialise output
    all_visible = []

    for i in range(L-1):
        node_visible = []
        ya = series[i]
        ta = timeLine[i]
        for j in range(i+1,L):

            yb = series[j]
            tb = timeLine[j]

            yc = series[i+1:j]
            tc = timeLine[i+1:j]

            if all( yc[k] < min(ya,yb) for k in range(len(yc)) ):
                node_visible.append(tb)
            elif any( yc[k] >= max(ya,yb) for k in range(len(yc)) ):
                break

        if len(node_visible)>0 : all_visible.append([ta, node_visible])

    return all_visible
```

Like natural visibility algorithm is the divide and conquer (D&C) approach for implementing the horizontal visibility algorithm that again taken from the same repository and changed at line 24 which makes it faster. I think it was a very minor error of coding in their code.

```python
def hvg_dc(series,timeLine, left, right, all_visible = None):

    if all_visible == None : all_visible = []

    node_visible = []

    if left < right :
        k = series[left:right].index(max(series[left:right])) + left

        for i in range(left,right):
            if i != k :
                a = min(i,k)
                b = max(i,k)

                ya = series[a]
                ta = timeLine[a]
                yb = series[b]
                tb = timeLine[b]
                yc = series[a+1:b]
                tc = timeLine[a+1:b]

                if all( yc[k] < min(ya,yb) for k in range(len(yc)) ):
                    node_visible.append(timeLine[i])
                elif any( yc[k] >= max(ya,yb) for k in range(len(yc)) ):
                    break

        if len(node_visible) > 0 : all_visible.append([timeLine[k], node_visible])

        hvg_dc(series,timeLine, left, k, all_visible = all_visible)
        hvg_dc(series,timeLine, k+1, right, all_visible = all_visible)

    return all_visible
```

```python
def graph_general_info(G):
    
    # some general information about the first heartbeat visibility graph
    print(nx.info(G))       
    # number of connected components
    print("\nNumber of connected components = ", nx.number_connected_components(G))

    # Top five nodes with the highest degree (In and Out degree)
    print("\n\nTop five nodes with the highest degree (In and Out degree)")
    a = sorted(dict(nx.degree(G)).items(), key=lambda x: x[1], reverse=True)  
    for node_id, node_degree in a[:5]:
        print("Node with sample number:", node_id, " => Degree: ", node_degree )
    return
# graph_general_info(hbs_vg[0])
```

```python
def plot_all_plots_single_hb(heart_beats, 
                             hbs_mov_avg, 
                             hbs_bpm_signal, 
                             hbs_annotations, 
                             bpm_list,
                             hbs_vg, 
                             hbs_hvg,
                             hb_id = 0):
    
    heart_beat = heart_beats[hb_id]
    hb_mov_avg = hbs_mov_avg[hb_id]
    hb_bpm_signal = hbs_bpm_signal[hb_id]
    hb_annotations = hbs_annotations[hb_id]
    bpm = bpm_list[hb_id]
    VG = hbs_vg[hb_id]
    HVG = hbs_hvg[hb_id]
    
    fig, ax_col_3 = plt.subplots(nrows=3, ncols=1, figsize=(17,9))
    
    top_ecg_analysis_plot(heart_beat, hb_mov_avg, hb_bpm_signal, hb_annotations, bpm, ax_col_3[0])
    plot_single_hb_vg(heart_beat, VG, ax_col_3[1])
    plot_single_hb_hvg(heart_beat, HVG, ax_col_3[2])
    
    plt.suptitle("Heart Beat number ="+str(hb_id+1), size=16)
    plt.show()
```

```python
def top_ecg_analysis_plot(heart_beat, hb_mov_avg, hb_bpm_signal, hb_annotations, bpm, ax):
    #Plot semi-transparent ECG
    ax.plot(heart_beat.tolist(), alpha=0.5, color='blue', ls='-', label="ECG")
    
    # set the labels to the exact sample number instead of starting from zero
    xlabels = np.array(ax.get_xticks().tolist()) + heart_beat.index.min()
    xlabels = xlabels.astype(int).tolist()
    ax.set_xticklabels(xlabels)
    
    #Plot simple moving average
    ax.plot(hb_mov_avg, color ='green', ls='-', label="SMA")

    # instantiate a second axes that shares the same x-axis
    ax2 = ax.twinx()
    
    #Plot bpm (beat per minutes for all the beats)
    ax2.plot(hb_bpm_signal, color ='purple', ls='-', label="BPM: %.1f " %bpm)
    ax.plot(np.nan, color ='purple', ls='-', label="BPM: %.1f " %bpm)
    
    # Add annotations
    xmin, xmax, ymin, ymax = ax2.axis()
    bbox_props1 = dict(boxstyle="round", fc="g", color="w", ec="0.5", alpha=0.5)
    bbox_props2 = dict(boxstyle="round", fc="darkred", color="w", ec="0.5", alpha=0.9)
    for index, annot in hb_annotations.iterrows():
        # set the annotation point (x,y) cordinate and text(x,y) cordinate
        annotation_text = annot['Type']
        annotation_text_x = annot['Sample'] - heart_beat.index.min()
        annotation_text_y = ymax
        annotation_x = annot['Sample']- heart_beat.index.min()
        
        if (annotation_text == "N"):
            plt.axvline(x=annotation_x, ymin=0, ymax=1, alpha=0.6, color='green', linestyle='dashed')
            ax2.text(annotation_text_x, annotation_text_y, annotation_text, 
                ha="center", va="center", size=10, color='w', bbox=bbox_props1)
        else:
            plt.axvline(x=annotation_x, ymin=0, ymax=1, color='darkred', linestyle='dashed')
            ax2.text(annotation_text_x, annotation_text_y, annotation_text, 
                ha="center", va="center", size=10, color='w', bbox=bbox_props2)
    
    ax.set_xlabel('Sample #', fontsize=8)
    ax.set_ylabel('Millivolts (mV)', color='tab:blue', fontsize=8)
    ax.tick_params(axis='y', labelcolor='tab:blue')

    ax2.set_ylabel('BPM', color='tab:purple', fontsize=8)
    ax2.tick_params(axis='y', labelcolor='tab:purple')

    ax.legend(framealpha=0.6, loc=1, scatterpoints=3, prop={"size":8})
```

```python
def plot_single_hb_vg(ecg_signal, G, ax):
    # node_positions
    time = ecg_signal.index.tolist()
    values = ecg_signal.values.tolist()
    x_y = list(zip(time, values))
    pos_dict = dict(zip(time, x_y))
    
    ax.plot(ecg_signal, alpha=1, color='blue', ls='-', label="ECG")
    
    # Set the size of each node as the degree of the node (Only display nodes with degree greater that 20)
    degrees = nx.degree(G)
#     sizes = [(degrees[node]+1) * 5 for node in G.nodes()]
    sizes = [(degrees[node]+1) * 5 if degrees[node] >= 20 else 0 for node in G.nodes()]
    
    nx.draw_networkx_nodes(G,
                           pos=pos_dict,
                           node_size=sizes,
                           alpha=.3,
                           ax=ax,
                           label="Nodes with degree > 20")
    
    labels = {node: node if degrees[node] >= 20 else '' for node in G.nodes}
    nx.draw_networkx_labels(G,
                            pos=pos_dict,
                            labels=labels,
                            node_size=sizes,
                            font_size=12,
                            font_color='black',
                            ax=ax,
                            alpha=1)
    
    nnodes = G.number_of_nodes()
    avg_deg = sum(d for n, d in degrees) / float(nnodes)
    avg_deg_str = str(round(avg_deg, 2))
    
    nx.draw_networkx_edges(G,
                           pos=pos_dict,
                           width=.5,
                           alpha=.5,
                           ax=ax,
                           label="Average degree: " + avg_deg_str)

    ax.legend(framealpha=0.6, loc=1, prop={"size":8})
    return
```

```python
def plot_single_hb_hvg(ecg_signal, G, ax):
    # node_positions
    time = ecg_signal.index.tolist()
    values = ecg_signal.values.tolist()
    x_y = list(zip(time, values))
    pos_dict = dict(zip(time, x_y))
    
    ax.plot(ecg_signal, alpha=1, color='blue', ls='-', label="ECG")
    
    # Set the size of each node as the degree of the node (Only display nodes with degree greater that 20)
    degrees = nx.degree(G)
#     sizes = [(degrees[node]+1) * 5 for node in G.nodes()]
    sizes = [(degrees[node]+1) * 20 if degrees[node] >= 5 else 0 for node in G.nodes()]

    nx.draw_networkx_nodes(G,
                           pos=pos_dict,
                           node_size=sizes,
                           alpha=.3,
                           ax=ax,
                           label="Nodes with degree > 5")
    
    labels = {node: node if degrees[node] >= 5 else '' for node in G.nodes}
    nx.draw_networkx_labels(G,
                            pos=pos_dict,
                            labels=labels,
                            node_size=sizes,
                            font_size=12,
                            font_color='black',
                            ax=ax,
                            alpha=1)
    
    nnodes = G.number_of_nodes()
    avg_deg = sum(d for n, d in degrees) / float(nnodes)
    avg_deg_str = str(round(avg_deg, 2))
    
    nx.draw_networkx_edges(G,
                           pos=pos_dict,
                           width=.075,
                           alpha=.5,
                           ax=ax,
                           label="Average degree: " + avg_deg_str)

    ax.legend(framealpha=0.6, loc=1, prop={"size":8})
    return
```

```python
def visibility_graph_hb(data_df, annotations_df, fs = 360, hb_id = 0):
    original_ecg = data_df[data_df.columns[1]]
    mov_avg = movingaverage(original_ecg, hrw=0.75)
    
    # test some possible ma_perc_shift for selecting the best from by looking at RRSD and BPM
    ma_perc_best = find_best_ma_perc_shift(original_ecg, mov_avg, fs)
    
    #Detect peaks with 'ma_perc_best'
    (peaklist, ybeat) = detect_peaks(original_ecg, mov_avg, ma_perc = ma_perc_best)
    RR_list = measure_rr(peaklist)
    bpm_list = measure_bpm(RR_list)
    (bpm_signal, avg_bpm) = measure_bpm_signal(mov_avg, bpm_list, peaklist)
    
    (heart_beats, hbs_mov_avg, hbs_bpm_signal, hbs_annotations) = hb_split(original_ecg, peaklist, mov_avg, 
                                                                           bpm_signal, annotations_df)

    hbs_vg = {}
    hbs_hvg = {}
    for hb_num, hb in enumerate(heart_beats):
        series = hb.tolist()
        time_samples = hb.index.values.tolist()
        nvg_array = nvg(series, time_samples)
        hvg_array = hvg(series, time_samples)
        
        hbs_vg[hb_num] = nx.Graph()
        for node_connections in nvg_array:
            edges = [(node_connections[0], item) for item in node_connections[1]]
            hbs_vg[hb_num].add_edges_from(edges)
        
        hbs_hvg[hb_num] = nx.Graph()
        for node_connections in hvg_array:
            edges = [(node_connections[0], item) for item in node_connections[1]]
            hbs_hvg[hb_num].add_edges_from(edges)        
    
    plot_all_plots_single_hb(heart_beats, 
                             hbs_mov_avg, 
                             hbs_bpm_signal, 
                             hbs_annotations, 
                             bpm_list,
                             hbs_vg, 
                             hbs_hvg,
                             hb_id = hb_id)
    print("\n------------------------Visibility Graph info HB =", hb_id+1 ,"------------------------")
    graph_general_info(hbs_vg[hb_id])
    print("\n------------------Horizontal Visibility Graph info HB =", hb_id+1 ,"-------------------")
    graph_general_info(hbs_hvg[hb_id])
    return hbs_vg, hbs_hvg, heart_beats

data_df = read_data_df(name='100', nRowsRead=None)
annotations_df = read_annotation_df(name='100')

(hbs_vg, hbs_hvg, heart_beats) = visibility_graph_hb(data_df, annotations_df, fs = 360, hb_id = 2)

print("total number of (heart_beats) = ", len(heart_beats))
```
